# **PROYECTO: SISTEMA DE GESTIÓN DE DATOS CON PYTHON**

Una vez visto el ***Módulo 1: Fundamentos de Programación***, se presenta la siguiente actividad para que el alumnado pueda poner a prueba lo aprendido durante este módulo.

Dentro de la siguiente dinámica, realizar un **menú CRUD** donde el tema a emplear es libre. Cada función deberá ejecutar una **operación del CRUD**, agregando una opción de **visualización**:

*   **Ver todos los registros**
*   **Crear nuevo registro**
*   **Consultar registro**
*   **Actualizar registro**
*   **Eliminar registro**
*   **Salir menú**

```python
inventario = []
def ver_registros():

    if len(inventario) == 0:
        print("No hay productos registrados")
        return

    print("\n--- Inventario completo ---")

    for i, producto in enumerate(inventario):
        print(f"\nProducto {i+1}")

        for key, value in producto.items():
            print(f"{key}: {value}")
def crear_registro():

    nombre = input("Nombre del producto: ")
    precio = input("Precio del producto: ")
    cantidad = input("Cantidad disponible: ")

    producto = {
        "Nombre": nombre,
        "Precio": precio,
        "Cantidad": cantidad
    }

    inventario.append(producto)

    print("Producto agregado correctamente")

def consultar_registro():

    buscar = input("Ingresa el nombre del producto a buscar: ")

    for producto in inventario:

        if producto["Nombre"] == buscar:
            print("Producto encontrado:")
            print(producto)
            return

    print("Producto no encontrado")

def actualizar_registro():

    buscar = input("Ingresa el nombre del producto a actualizar: ")

    for producto in inventario:

        if producto["Nombre"] == buscar:

            while True:

                print("\nMenú de actualización")
                print("1. Cambiar nombre")
                print("2. Cambiar precio")
                print("3. Cambiar cantidad")
                print("4. Salir")

                opcion = input("Selecciona una opción: ")

                if opcion == "1":
                    producto["Nombre"] = input("Nuevo nombre: ")
                    print(producto)

                elif opcion == "2":
                    producto["Precio"] = input("Nuevo precio: ")
                    print(producto)

                elif opcion == "3":
                    producto["Cantidad"] = input("Nueva cantidad: ")
                    print(producto)

                elif opcion == "4":
                    break

                else:
                    print("Opción no válida")

            return

    print("Producto no encontrado")

def eliminar_registro():

    buscar = input("Ingresa el nombre del producto a eliminar: ")

    for producto in inventario:

        if producto["Nombre"] == buscar:
            inventario.remove(producto)
            print("Producto eliminado")
            return

    print("Producto no encontrado")

def menu():

    while True:

        print("\n--- MENÚ CRUD INVENTARIO ---")
        print("1. Ver todos los registros")
        print("2. Crear nuevo registro")
        print("3. Consultar registro")
        print("4. Actualizar registro")
        print("5. Eliminar registro")
        print("6. Salir")

        opcion = input("Selecciona una opción: ")

        if opcion == "1":
            ver_registros()

        elif opcion == "2":
            crear_registro()

        elif opcion == "3":
            consultar_registro()

        elif opcion == "4":
            actualizar_registro()

        elif opcion == "5":
            eliminar_registro()

        elif opcion == "6":
            print("Saliendo del sistema...")
            break

        else:
            print("Opción no válida")

menu()
```
